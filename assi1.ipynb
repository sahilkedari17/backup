```python
print(" ChatBot: Hello! I am your simple chatbot.")
print("Type 'bye' to exit.\n")

def chatbot_response(user_input):
    user_input = user_input.lower()

    responses = {
        "hi": "Hello there!",
        "hello": "Hi! How can I help you?",
        "how are you": "I'm just a program, but I'm doing great!",
        "what is your name": "My name is SimpleBot.",
        "who created you": "I was created by a Python programmer.",
        "what can you do": "I can chat with you using simple rules.",
        "bye": "Goodbye! Have a nice day!"
    }

    for key in responses:
        if key in user_input:
            return responses[key]

    return "Sorry, I don't understand that."


while True:
    user = input("You: ")
    if user.lower() == "bye":
        print(" ChatBot: Goodbye! Have a nice day!")
        break

    reply = chatbot_response(user)
    print(" ChatBot:", reply)
```

```python
from collections import deque

def bfs(graph, start):
    visited = set()         
    queue = deque()          

    visited.add(start)
    queue.append(start)

    print("BFS Traversal:", end=" ")

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

start_node = 'A'
bfs(graph, start_node)
```

```python
def dfs(graph, node, visited):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)

        for neighbour in graph[node]:
            dfs(graph, neighbour, visited)


graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

visited = set()
start_node = 'A'

print("DFS Traversal:", end=" ")
dfs(graph, start_node, visited)
```

```python
from collections import deque


CAP_A = 4
CAP_B = 3
TARGET = 2

def is_valid(state):
    a, b = state
    return 0 <= a <= CAP_A and 0 <= b <= CAP_B

def get_next_states(state):
    a, b = state
    states = []

    
    states.append((CAP_A, b))

    
    states.append((a, CAP_B))

    
    states.append((0, b))

    
    states.append((a, 0))

    
    pour = min(a, CAP_B - b)
    states.append((a - pour, b + pour))

    
    pour = min(b, CAP_A - a)
    states.append((a + pour, b - pour))

    return states

def water_jug_bfs():
    start = (0, 0)
    queue = deque()
    queue.append((start, [start]))
    visited = set()
    visited.add(start)

    while queue:
        (a, b), path = queue.popleft()

        
        if a == TARGET or b == TARGET:
            print("Solution Path:")
            for step in path:
                print(step)
            return

        for state in get_next_states((a, b)):
            if state not in visited and is_valid(state):
                visited.add(state)
                queue.append((state, path + [state]))

    print("No solution found.")

water_jug_bfs()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("iris.csv")


setosa = data[data['species'] == 'Iris-setosa']
versicolor = data[data['species'] == 'Iris-versicolor']
virginica = data[data['species'] == 'Iris-virginica']


plt.figure()
plt.scatter(setosa['sepal_length'], setosa['sepal_width'])
plt.scatter(versicolor['sepal_length'], versicolor['sepal_width'])
plt.scatter(virginica['sepal_length'], virginica['sepal_width'])
 
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset Scatter Plot")

plt.show()
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer


data = pd.read_csv("iris.csv")

print("Null values before imputation:\n")
print(data.isnull().sum())

numeric_data = data.select_dtypes(include=['float64', 'int64'])


imputer = SimpleImputer(strategy='mean')


data[numeric_data.columns] = imputer.fit_transform(numeric_data)

print("\nNull values after imputation:\n")
print(data.isnull().sum())

data.to_csv("iris_cleaned.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


data = pd.read_csv("iris.csv")


label_encoder = LabelEncoder()
data['species_label'] = label_encoder.fit_transform(data['species'])

print("Label Encoded Data:\n")
print(data[['species', 'species_label']].head())


onehot_encoder = OneHotEncoder(sparse_output=False)

encoded_data = onehot_encoder.fit_transform(data[['species']])
encoded_df = pd.DataFrame(encoded_data, columns=onehot_encoder.get_feature_names_out(['species']))


data = pd.concat([data, encoded_df], axis=1)

print("\nOne-Hot Encoded Data:\n")
print(encoded_df.head())


data.to_csv("iris_encoded.csv", index=False)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler


data = pd.read_csv("Salary_Data.csv")

print("Before Standardization:\n")
print(data.head())


numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns


scaler = StandardScaler()
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

print("\nAfter Standardization:\n")
print(data.head())


data.to_csv("Salary_Data_Standardized.csv", index=False)
```

```python
# House Price Prediction using Linear Regression
# Dataset: House_price_prediction.csv

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
data = pd.read_csv("Housing.csv")

print("Columns in Dataset:")
print(data.columns)

# ----------------------------------------------------
# 1. SIMPLE LINEAR REGRESSION (area)
# ----------------------------------------------------

# Feature and target
X_simple = data[['area']]     # Independent variable
y = data['price']             # Dependent variable

# Train-test split
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X_simple, y, test_size=0.2, random_state=42
)

# Model
simple_lr = LinearRegression()
simple_lr.fit(X_train_s, y_train_s)

# Prediction
y_pred_s = simple_lr.predict(X_test_s)

# Evaluation
print("\n--- Simple Linear Regression ---")
print("Coefficient:", simple_lr.coef_[0])
print("Intercept:", simple_lr.intercept_)
print("MSE:", mean_squared_error(y_test_s, y_pred_s))
print("R2 Score:", r2_score(y_test_s, y_pred_s))

# Plot
plt.scatter(X_test_s, y_test_s, color='blue', label="Actual Price")
plt.plot(X_test_s, y_pred_s, color='red', label="Predicted Price")
plt.xlabel("Area (sqft)")
plt.ylabel("House Price")
plt.title("Simple Linear Regression (Area vs Price)")
plt.legend()
plt.show()

# ----------------------------------------------------
# 2. MULTIPLE LINEAR REGRESSION
# ----------------------------------------------------

# Select numeric features only
features = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking']

X_multi = data[features]
y_multi = data['price']

# Train-test split
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi, y_multi, test_size=0.2, random_state=42
)

# Model
multi_lr = LinearRegression()
multi_lr.fit(X_train_m, y_train_m)

# Prediction
y_pred_m = multi_lr.predict(X_test_m)

# Evaluation
print("\n--- Multiple Linear Regression ---")
print("Coefficients:")
for feature, coef in zip(features, multi_lr.coef_):
    print(f"{feature}: {coef}")

print("Intercept:", multi_lr.intercept_)
print("MSE:", mean_squared_error(y_test_m, y_pred_m))
print("R2 Score:", r2_score(y_test_m, y_pred_m))
```

```python
# Polynomial Regression Implementation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Load dataset
data = pd.read_csv("Position_Salaries.xls")

print("Dataset Preview:")
print(data.head())

# Independent and dependent variables
X = data[['Level']]   # Feature must be 2D
y = data['Salary']

# -------------------------------
# Linear Regression Model
# -------------------------------
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# -------------------------------
# Polynomial Regression Model
# -------------------------------
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# -------------------------------
# Visualization
# -------------------------------

# Linear Regression plot
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red')
plt.title("Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Polynomial Regression plot (smooth curve)
X_grid = np.arange(min(X.values), max(X.values), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color='blue')
plt.plot(X_grid, poly_reg.predict(poly.fit_transform(X_grid)), color='red')
plt.title("Polynomial Regression (Degree = 4)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# -------------------------------
# Prediction Example
# -------------------------------
level = [[6.5]]

linear_salary = lin_reg.predict(level)
poly_salary = poly_reg.predict(poly.fit_transform(level))

print("\nPrediction for Level 6.5:")
print("Linear Regression Salary:", linear_salary[0])
print("Polynomial Regression Salary:", poly_salary[0])
```

```python
# Non-Linear Regression Models
# Decision Tree Regression & Support Vector Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

# Load dataset
data = pd.read_csv("Position_Salaries.xls")

print("Dataset Preview:")
print(data.head())

# Feature and target
X = data[['Level']].values
y = data['Salary'].values

# ----------------------------------------------------
# 1. DECISION TREE REGRESSION
# ----------------------------------------------------

dt_reg = DecisionTreeRegressor(random_state=0)
dt_reg.fit(X, y)

# Visualization (high resolution)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(-1, 1)

plt.scatter(X, y, color='blue')
plt.plot(X_grid, dt_reg.predict(X_grid), color='red')
plt.title("Decision Tree Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Prediction
dt_pred = dt_reg.predict([[6.5]])
print("\nDecision Tree Prediction for Level 6.5:", dt_pred[0])

# ----------------------------------------------------
# 2. SUPPORT VECTOR REGRESSION (SVR)
# ----------------------------------------------------

# Feature scaling (IMPORTANT for SVR)
sc_X = StandardScaler()
sc_y = StandardScaler()

X_scaled = sc_X.fit_transform(X)
y_scaled = sc_y.fit_transform(y.reshape(-1, 1)).ravel()

svr_reg = SVR(kernel='rbf')
svr_reg.fit(X_scaled, y_scaled)

# Visualization
plt.scatter(X_scaled, y_scaled, color='blue')
plt.plot(X_scaled, svr_reg.predict(X_scaled), color='red')
plt.title("Support Vector Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()

# Prediction
svr_pred_scaled = svr_reg.predict(sc_X.transform([[6.5]]))
svr_pred = sc_y.inverse_transform(svr_pred_scaled.reshape(-1, 1))

print("SVR Prediction for Level 6.5:", svr_pred[0][0])
```
